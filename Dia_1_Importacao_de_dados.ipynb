#Dia 1 - Importação de dados


```python
import pandas as pd
```

## Coletando dados

```python
dados_2010_01 = pd.read_csv('emprestimos-20101.csv')
dados_2010_02 = pd.read_csv('emprestimos-20102.csv')
dados_2011_01 = pd.read_csv('emprestimos-20111.csv')
dados_2011_02 = pd.read_csv('emprestimos-20112.csv')
dados_2012_01 = pd.read_csv('emprestimos-20121.csv')
dados_2012_02 = pd.read_csv('emprestimos-20122.csv')
dados_2013_01 = pd.read_csv('emprestimos-20131.csv')
dados_2013_02 = pd.read_csv('emprestimos-20132.csv')
dados_2014_01 = pd.read_csv('emprestimos-20141.csv')
dados_2014_02 = pd.read_csv('emprestimos-20142.csv')
dados_2015_01 = pd.read_csv('emprestimos-20151.csv')
dados_2015_02 = pd.read_csv('emprestimos-20152.csv')
dados_2016_01 = pd.read_csv('emprestimos-20161.csv')
dados_2016_02 = pd.read_csv('emprestimos-20162.csv')
dados_2017_01 = pd.read_csv('emprestimos-20171.csv')
dados_2017_02 = pd.read_csv('emprestimos-20172.csv')
dados_2018_01 = pd.read_csv('emprestimos-20181.csv')
dados_2018_02 = pd.read_csv('emprestimos-20182.csv')
dados_2019_01 = pd.read_csv('emprestimos-20191.csv')
dados_2019_02 = pd.read_csv('emprestimos-20192.csv')
dados_2020_01 = pd.read_csv('emprestimos-20201.csv')
```

## Concatenando os DataFrames

```python
emprestimos = pd.concat([dados_2010_01, dados_2010_02, dados_2011_01, 
                         dados_2011_02, dados_2012_01, dados_2012_02, 
                         dados_2013_01, dados_2013_02, dados_2014_01,
                         dados_2014_02, dados_2015_01, dados_2015_02,
                         dados_2016_01, dados_2016_02, dados_2017_01,
                         dados_2017_02, dados_2018_01, dados_2018_02,
                         dados_2019_01, dados_2019_02, dados_2020_01], 
                        ignore_index=True)

emprestimos
```

## Verificando duplicatas

```python
print(emprestimos.duplicated())
print(emprestimos.value_counts())
```

## Excluindo duplicatas

```python
print(emprestimos.drop_duplicates(inplace = True))
print(emprestimos.value_counts())
```

```python
# Índice ficou quebrado após a exclusão das duplicatas 
emprestimos.head()
```

```python
# Corrigindo o índice
emprestimos.reset_index(drop=True, inplace=True)
emprestimos.head()
```

## Importando mais dados

```python
dados_exemplares = pd.read_parquet('dados_exemplares.parquet')
dados_exemplares
```

## Unindo os DataFrames

```python
emprestimo_completo = emprestimos.merge(dados_exemplares)
emprestimo_completo.head()
```
